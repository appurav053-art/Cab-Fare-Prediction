Feature Scaling For (Train) Dataset

```python
for i in ['fare_amount', 'distance']:
    print(i)
    sns.distplot(train[i],bins='auto',color='green')
    
    plt.title("Distribution for Variable "+i)
    
    plt.ylabel("Density")
    plt.show()
```
